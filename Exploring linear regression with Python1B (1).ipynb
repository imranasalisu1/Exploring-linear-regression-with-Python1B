```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
data = pd.read_csv('penguins.csv')
data
```

```python
# DATA CLEANING
data_clean = data.dropna(subset=['bill_length_mm', 'flipper_length_mm','sex','year','body_mass_g','species','island'])
data_clean
```

```python
data_clean.describe()
```

```python
# CHECKING THE MISSING VALUES AGAIN
data_clean.isnull()
```

```python
std_bill_length_mm = data['bill_length_mm'].std()
std_bill_length_mm
```

```python
# Create pairwise scatterplots of data set
sns.pairplot(data_clean)
```

```python
# From the scatterplot matrix, we can observe a few linear relationships:
#1 bill length (mm) and flipper length (mm)
#2 bill length (mm) and body mass (g)
#3 flipper length (mm) and body mass (g)
```

```python
# Model construction
# Based on the above scatterplots, we could probably run a simple linear regression on any of the three relationships identified.
# For this part of the course, we will focus on the relationship between bill length (mm) and body mass (g).
# To do this, we will first subset the variables of interest from the dataframe.
# we can do this by using double square brackets [[]], and listing the names of the columns of interest.
```

```python
# Import ols function
from statsmodels.formula.api import ols
```

```python
# Then, you can plug in the ols_formula and ols_data as arguments in the ols() function.
# After you save the results as a variable, you can call on the fit() function to actually fit the model to the data.
```

```python
# Subset Data
ols_data = data_clean[["bill_length_mm", "body_mass_g"]]
ols_data
```

```python
# Import ols function
from statsmodels.formula.api import ols
import statsmodels.formula.api as smf
```

```python
# To build an OLS model using statsmodels, we need to define the formula that specifies the relationship,
# between the dependent and independent variables.
# The formula should be in the format dependent_variable ~ independent_variable(s).
```

```python
ols_formula = 'body_mass_g ~ bill_length_mm'
ols_formula
```

```python
# Build OLS, fit model to data
OLS = smf.ols(formula=ols_formula, data=ols_data)
model = OLS.fit()
```

```python
model.summary()
```

```python
# we can use the regplot() function from seaborn to visualize the regression line
```

```python
sns.regplot(x = "bill_length_mm", y = "body_mass_g", data = ols_data)
```

```python
# Interpretation of the OLS Output
# Coefficients: The estimated coefficients for each predictor and the intercept.
# R-squared: Indicates the proportion of the variance in the dependent variable that is predictable from the independent variable(s).
# P-values: Helps to determine the statistical significance of each coefficient. A p-value less than 0.05 typically indicates that the predictor is significant.
# F-statistic: Tests the overall significance of the model. A higher F-statistic with a low p-value indicates that the model is statistically significant.
# Durbin-Watson: Tests for autocorrelation in the residuals. A value close to 2 suggests no autocorrelation.
# Prob (F-statistic): The p-value for the F-test of the overall model fit. A low value indicates that the model is statistically significant.
```

```python
# CHECKING THE MODEL ASSUMPTIONS

#a The normality and homoscedasticity assumptions focus on the distribution of errors
#b Thus, we can only check these assumptions after we have constructed the model.
#c To check these assumptions, you will check the residuals, as an approximation of the errors.
#d To more easily check the model assumptions and create relevant visualizations,
#e we can first subset the X variable by isolating just the bill_length_mm column.
#f  Additionally, we can save the predicted values from the model using the model.predict(X) function.
```

```python
# Subset X variable
X = ols_data["bill_length_mm"]
X

# Get predictions from model
fitted_values = model.predict(X)
fitted_values
# Then, you can save the model residuals as a variable by using the model.resid attribute.
```

```python
# Calculate residuals
residuals = model.resid
residuals
```

```python
# Check the normality assumption

#1 From the below histogram, we may notice that the residuals are almost normally distributed.
#2 In this case, it is likely close enough that the assumption is met.
```

```python
import matplotlib.pyplot as plt
fig = sns.histplot(residuals)
fig.set_xlabel("Residual Value")
fig.set_title("Histogram of Residuals")
plt.show()
```

```python
# Another way to check the normality function is to create a quantile-quantile or Q-Q plot.
#i. Recall that if the residuals are normally distributed, we would expect a straight diagonal line going from the bottom left to the upper right of the Q-Q plot. 
#ii.we can create a Q-Q plot by using the qqplot function from the statsmodels.api package.
#iii.The Q-Q plot shows a similar pattern to the histogram, where the residuals are mostly normally distributed, except at the ends of the distribution.
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
fig = sm.qqplot(model.resid, line = 's')
plt.show()
```

```python
# Check the homoscedasticity assumption
#a.Lastly, we have to check the homoscedasticity assumption.
#b.To check the homoscedasticity assumption, you can create a scatterplot of the fitted values and residuals.
#c.If the plot resembles a random cloud (i.e., the residuals are scattered randomly), then the assumption is likely met.

#You can create one scatterplot by using the scatterplot() function from the seaborn package.
# The first argument is the variable that goes on the x-axis. The second argument is the variable that goes on the y-axis.
```

```python
# Import matplotlib
import matplotlib.pyplot as plt
fig = sns.scatterplot(x=fitted_values, y=residuals)

# Add reference line at residuals = 0
fig.axhline(0)

# Set x-axis and y-axis labels
fig.set_xlabel("Fitted Values")
fig.set_ylabel("Residuals")

# Show the plot
plt.show()
```
